```python
from sklearn.cluster import KMeans

"""
n_clusters: the number of desired clusters

n_init: run the k-means clustering algorithms 10 times independently with different 
        random centroids to choose the final model as the one with the lowest SSE
        
max_iter: maximum number of iterations for each single run

tol: tolerance with regard to the changes in the within-cluster SSE to declare convergence

"""

km = KMeans(n_clusters = 2, init = 'random', n_init = 20, max_iter =300, random_state = 0)
y_km = km.fit_predict(x_pca)
y_km
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the 2 clusters
plt.scatter(x_pca[y_km == 0, 0], x_pca[y_km == 0, 1], s = 50, c = 'lightgreen', marker = 's',label ='Cluster 1')

plt.scatter(x_pca[y_km == 1, 0], x_pca[y_km == 1, 1], s = 50, c = 'orange', marker = 'o', label = 'Cluster 2')

# plot the centroids
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 700, marker ='.', c ='red', label = 'Centroids')

plt.legend(scatterpoints = 1)
plt.savefig('Cluster')
plt.show()
```

```python

distortions = []
for i in range(1, 15):
    km = KMeans(n_clusters = i, init = 'random', n_init = 20, max_iter = 300, tol = 1e-04, random_state = 40)
    km.fit(x_pca)
    distortions.append(km.inertia_)

    
# plot
plt.plot(range(1, 15), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.savefig('kmeans')
plt.show()
```
